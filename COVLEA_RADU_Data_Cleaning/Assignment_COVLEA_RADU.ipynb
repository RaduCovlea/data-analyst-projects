> Voi incepe acest proiect prin a importa package-ul necesar realizarii lui. Am considerat ca singurul package necesar este "pandas"

```python
import pandas as pd
```

> Am incarcat seturile de date, am folosit parametrul 'on_bad_lines' pentru a evita erorile asociate lipsei de date si neprelucrarii datelor

```python
facebook_data = pd.read_csv('C:\\Users\\user\\Desktop\\datasets\\facebook_dataset.csv', on_bad_lines='skip')
google_data = pd.read_csv('C:\\Users\\user\\Desktop\\datasets\\google_dataset.csv', on_bad_lines='skip', low_memory=False)
website_data = pd.read_csv('C:\\Users\\user\\Desktop\\datasets\\website_dataset.csv', sep=';', on_bad_lines='skip')
```

```python
print(facebook_data.columns)
print(google_data.columns)
print(website_data.columns)
```

> In ceea ce urmeaza am redenumit coloanele comune pentru uniformitate si pentru a ma putea referi la ele ulterior

```python
facebook_subset = facebook_data[['name','address','city','region_name','country_name', 'domain', 'categories', 'phone']]
facebook_subset.columns = ['company_name', 'address', 'city', 'region', 'country', 'domain', 'category', 'phone']
facebook_subset.head()

google_subset = google_data[['name', 'address', 'city', 'region_name', 'country_name', 'domain', 'category', 'phone']]
google_subset.columns = ['company_name', 'address', 'city', 'region', 'country', 'domain', 'category', 'phone']
google_subset.head()

website_subset = website_data[['legal_name', 'main_city', 'main_region', 'main_country', 'root_domain', 's_category', 'phone']]
website_subset.columns = ['company_name', 'city', 'region', 'country', 'domain', 'category', 'phone']
website_subset.head()
```

> Am combinat seturile de date folosing ca si punct fix pentru join coloana "company_name", intrucat este esentiala si orice firma este obligatoriu sa o aiba. Daca nu exista numele companiei, atunci firma trebuie exclusa din dataset sau sa organizam o munca titanica facand research cu informatiile pe care le avem la dispozitie, dar nu cred ca este cazul :) De asemenea, am adaugat sufixe pentru a stii carui dataset ii corespunde fiecare valoare. Este util in faza preliminara prelucrarii datelor, insa pentru dataset-ul de websites am considerat ca ii voi asocia un sufix doar daca exista conflict de date.

```python
merged_companies = pd.merge(facebook_subset, google_subset, on='company_name', how='outer', suffixes=('_fb', '_gg'))
merged_companies = pd.merge(merged_companies, website_subset, on='company_name', how='outer', suffixes=('', '_wb'))
merged_companies.sample(10)
```

```python
print(merged_companies.columns)
```

> Am creat o copie a dataframe-ului curent pentru a avea un checkpoint la care sa ma intorc in cazul in care voi gresi, ceea ce s-a si intamplat :)

```python
merged_companies_v1 = merged_companies.copy()
merged_companies_v1.head()
```

> Pentru a rezolva conflictele de date am ales sa completez cu prima valoare non-null pe care o gasesc asociata fiecarei coloane si observatii. Am considerat aceasta solutie ca fiind fiabila, presupunand ca dataset-urile nu contin informatii gresite si pentru ca nu a fost prezentat in descrierea proiectului daca unul dintre dataset-uri este mai trustworthy decat altul.

> Coloana "company_name" nu a fost inclusa, deoarece riscam valori duplicate si alte probleme in urma utilizarii acesteia pentru merging

```python
merged_companies_v1['address'] = merged_companies['address_fb'].combine_first(merged_companies['address_gg'])
merged_companies_v1['city'] = merged_companies['city_fb'].combine_first(merged_companies['city_gg']).combine_first(merged_companies['city'])
merged_companies_v1['region'] = merged_companies['region_fb'].combine_first(merged_companies['region_gg']).combine_first(merged_companies['region'])
merged_companies_v1['country'] = merged_companies['country_fb'].combine_first(merged_companies['country_gg']).combine_first(merged_companies['country'])
merged_companies_v1['domain'] = merged_companies['domain_fb'].combine_first(merged_companies['domain_gg']).combine_first(merged_companies['domain'])
merged_companies_v1['category'] = merged_companies['category_fb'].combine_first(merged_companies['category_gg']).combine_first(merged_companies['category'])
merged_companies_v1['phone'] = merged_companies['phone_fb'].combine_first(merged_companies['phone_gg']).combine_first(merged_companies['phone'])
```

> Am renuntat acum la coloanele cu sufix, intrucat am populat coloanele care vor face obiectul proiectului cu informatie

```python
merged_companies_v1 = merged_companies_v1.drop(columns=['address_fb','city_fb','region_fb','country_fb','domain_fb','category_fb','phone_fb',
                              'address_gg','city_gg','region_gg','country_gg','domain_gg','category_gg','phone_gg'])
```

Observam ca datatype-ul pentru fiecare coloana este corect, neavand valori explicit numerice, insa utilizarea memoriei este foarte mare din cauza neprelucrarii datelor

> In cazul in care exista observatii care au informatii identice in coloanele "company_name", "phone" si "address" am renuntat la ele.

> Bineinteles, dupa ce m-am asigurat ca am un checkpoint, decizia mea a fost sa inlatur observatiile care au acelasi numar de telefon si aceeasi adresa, intrucat dupa ce am analizat dataset-ul mi-am dat seama ca o companie poate sa aiba acelasi nume, insa multiple sedii care se afla in orase sau regiuni diferite. Chiar si asa, fiecare sediu al companiei ar trebui sa aiba un numar de telefon si o adresa de sine statatoare. Ca si adaugare, am ales sa izolez valorile nule de care avem nevoie si sa pastrez prima valoare in cazul duplicatelor din aceleasi motive ca mai sus, nu este specificat daca un dataset are prioritate in considerarea alegerii informatiilor fata de altul.

```python
nan_phone = merged_companies_v1[merged_companies_v1['phone'].isna()]
merged_companies_v1_cleaned = merged_companies_v1.dropna(subset=['phone']).drop_duplicates(subset=['phone'], keep='first')
merged_companies_v1 = pd.concat([merged_companies_v1_cleaned, nan_phone])

nan_address = merged_companies_v1[merged_companies_v1['address'].isna()]
merged_companies_v1_cleaned = merged_companies_v1.dropna(subset=['address']).drop_duplicates(subset=['address'], keep='first')
merged_companies_v1 = pd.concat([merged_companies_v1_cleaned, nan_address])

merged_companies_v1 = merged_companies_v1.reset_index(drop=True)
```

> Am folosit o functie pentru a inlocui toate valorile nule cu "Unknown" pentru ca dataset-ul sa fie usor de inteles pentru oricine il citeste. Lizibilitatea este un factor important al datelor, asta am fost invatat in cursul de Data Analyst pe care l-am absolvit.

```python
def replace_nan_with_unknown(df):
    return df.fillna('Unknown')

merged_companies_v1 = replace_nan_with_unknown(merged_companies_v1)
```

> Am vrut sa vad si o suma a valorilor "Unknown" si aparent nu este asa infricosator avand in vedere numarul de observatii pe care le avem.

```python
print(merged_companies_v1.apply(lambda x: (x == 'Unknown').sum()))
```

> Am observat cea mai problematica coloana, si anume "phone" care nu avea un format uniform. Numerele de telefon erau corecte, stiu asta pentru ca am facut research, insa nu apareau cu acel semn de "+" in fata care indica sufixul aferent fiecarei tari si l-am adaugat tot prin folosirea unei functii in cazul in care acesta nu se regaseste deja.

> Alta problema a fost terminatia unor valori ale acestei coloane cu ".0", insa cum datele nu sunt numerice de tip float sau double am folosit si aici o functie pentru a preschimba datele in tipul string, deoarece primeam o eroare care spunea ca trebuie sa schimb datatype-ul pentru a putea face asemenea operatii (din cauza valorilor "Unknown") si scopul principal al functiei este de a renunta la acele terminatii specifice valorilor numerice.

```python
merged_companies_v1['phone'] = merged_companies_v1['phone'].apply(lambda x:x if x.startswith('+') else f'+{x}')
merged_companies_v1['phone'] = merged_companies_v1['phone'].astype(str).apply(lambda x: x.rstrip('.0') if x.endswith('.0') else x)
```

```python
merged_companies_v1.sample(20)
```

> Am mai folosit o functie pentru a elimina "+" din valorile "Unknown", aceasta problema fiind o consecinta a functiei de mai sus.

```python
merged_companies_v1['phone'] = merged_companies_v1['phone'].apply(lambda x: "Unknown" if x=='+Unknown' else x)
```

> Pentru o scriere corecta si o lizibilitate mai buna am asociat uppercase tuturor cuvintelor care se afla in coloanele "city", "region" si "country". In coloana "address" nu erau probleme.

```python
merged_companies_v1['city'] = merged_companies_v1['city'].str.title()
merged_companies_v1['region'] = merged_companies_v1['region'].str.title()
merged_companies_v1['country'] = merged_companies_v1['country'].str.title()
```

Informatiile sunt identice ca si cele din dataset si prin prelucrare obtinem o imbunatatire evidenta, nu mai exista valori nule, memoria folosita este mult mai mica, am scapat de duplicated si acum avem un dataset mult mai curat.

> Prin urmatoarea functie m-am asigurat ca putem afisa valorile coloanei "problematice" asa cum trebuie, intrucat mie imi apareau fara acel "+" cand deschideam fisierul csv asa ca am cautat cum sa rezolv si o solutie fiabila mi-a parut sa adaug un apostrof la inceputul fiecarei observatii.

```python
merged_companies_v1['phone'] = merged_companies_v1['phone'].apply(lambda x: f"'{x}" if x.startswith('+') else x)
```
