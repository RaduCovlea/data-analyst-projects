## 3. Clean data

```python
# FILL IN - Make copies of the datasets to ensure the raw dataframes 
# are not impacted
world_cup_m = world_cup_matches.copy()
players = pl_players.copy()
```

### **Quality Issue 1: MISSING VALUES**

```python
# FILL IN - Apply the cleaning strategy
# We will drop the NA values so we solve the completeness issue
world_cup_m.dropna(inplace=True)
players.dropna(inplace=True)
```

```python
# FILL IN - Validate the cleaning was successful
print(world_cup_m.info())
print(players.info())
```

Justification: After we drop the NA values we can observe that data is more compact and the memory usage decreased.

### **Quality Issue 2: Inconsistent data types**

```python
#FILL IN - Apply the cleaning strategy
#we will modify datatypes of some variables
world_cup_m['Year'] = world_cup_m['Year'].astype(int)
world_cup_m['Home Team Goals'] = world_cup_m['Home Team Goals'].astype(int)
world_cup_m['Away Team Goals'] = world_cup_m['Away Team Goals'].astype(int)
world_cup_m[['Date', 'Time']] = world_cup_m['Datetime'].str.split(' - ', expand=True)
date_formats = ['%d %b %Y', '%d %B %Y']
for date_format in date_formats:
    world_cup_m['Date'] = pd.to_datetime(world_cup_m['Date'], format=date_format, errors='coerce')
world_cup_m.drop(columns=['Datetime'], inplace=True)
```

```python
#FILL IN - Validate the cleaning was successful
world_cup_m.info()
world_cup_m.head(10)
```

Justification: After we changed the datatypes we see that 'Year' and goals variables are now displayed correct and we can perform analysis on 'Date' variable now that it is stored as datetime object

### **Tidiness Issue 1: New 'Winner' column**

```python
#FILL IN - Apply the cleaning strategy
#let's see which team won every game by counting the number of goals they scored versus their competitors
def determine_winner(row):
    if row['Home Team Goals'] > row['Away Team Goals']:
        return row['Home Team Name']
    elif row['Home Team Goals'] < row['Away Team Goals']:
        return row['Away Team Name']
    else:
        return "Draw"

# adding the new column 'Winner'
world_cup_m['Winner'] = world_cup_m.apply(determine_winner, axis=1)
```

```python
#FILL IN - Validate the cleaning was successful
world_cup_m.head(40)
```

Justification: After we added this new 'Winner' column now we know the result of every game from the World Cup and we are one step closer to response the research question.

### **Tidiness Issue 2: Split 'Name' variable**

```python
#FILL IN - Apply the cleaning strategy
players[['First Name', 'Last Name']] = players['Name'].str.split(' ', n=1, expand=True)
players.drop(columns=['Name'],inplace=True)
```

```python
#FILL IN - Validate the cleaning was successful
players.head()
```

Justification: After we split 'Name' variable we can now perform more detailed analysis for every name and every player.

### **Remove unnecessary variables and combine datasets**

Depending on the datasets, you can also peform the combination before the cleaning steps.

```python
#FILL IN - Remove unnecessary variables and combine datasets
# adding the new column
world_cup_m['Competition Winner'] = "Watch for the final"

# identifying the final matches and updating the column accordingly
final_matches = world_cup_m[world_cup_m['Stage'] == 'Final']

# update the 'Competition Winner' for each final match winner
for index, match in final_matches.iterrows():
    winner = match['Home Team Name'] if match['Home Team Goals'] > match['Away Team Goals'] else match['Away Team Name']
    year = match['Year']
    world_cup_m.loc[world_cup_m['Year'] == year, 'Competition Winner'] = world_cup_m.apply(
        lambda row: winner if row['Stage'] == 'Final' else "Watch for the final",
        axis=1
    )
```

```python
world_cup_m.head(40)
```

```python
world_cup_m['Competition Winner'].value_counts()
```

```python
# merging on Home Team Name
merged_home_national = pd.merge(world_cup_m, pl_players, left_on="Home Team Name", right_on="Nation", how="left")

# merging on Away Team Name
merged_away_national = pd.merge(world_cup_m, pl_players, left_on="Away Team Name", right_on="Nation", how="left")

#print("Merged Data (Home Team):")
#print(merged_home_national.head())
#print("\nMerged Data (Away Team):")
#print(merged_away_national.head())

# combining the two merged DataFrames
combined_merged = pd.concat([merged_home_national, merged_away_national])

# displaying the combined DataFrame
print("Combined Merged Data:")
combined_merged.head()
```

### **5.1:** Define and answer the research question 
Going back to the problem statement in step 1, use the cleaned data to answer the question you raised. Produce **at least** two visualizations using the cleaned data and explain how they help you answer the question.

*Research question:* Which many players are right now in Premier League representing nationalities than won 3 World Cups or more
   
       Above, we seen that there are 3 national teams who won 3 WC or more: Italy (4), Brazil(3), Germany(3). So, we filter our data to display only informations about these player's nationalities.

```python
#Visual 1 - FILL IN
import matplotlib.pyplot as plt
filtered_nation = combined_merged[combined_merged['Nation'].isin(['Italy', 'Brazil', 'Germany'])]

nation_counts = filtered_nation['Nation'].value_counts()

plt.figure(figsize=(10, 6))
nation_counts.plot(kind='bar')
plt.title('Number of Players by Nations who won more than 3 World Cups')
plt.xlabel('Nation')
plt.ylabel('Number of Players')
plt.xticks(rotation=0)
plt.show()
```

*Answer to research question:* We observe that there are a huge amount of Brazilian players in the England league and about 50 german players, but Italy got 0 players in this league.

```python
#Visual 2 - FILL IN
plt.figure(figsize=(10, 6))
nation_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Players by Nations who won more than 3 World Cups')
plt.ylabel('') 
plt.show()
```

*Answer to research question:* We observe that there are a huge amount of Brazilian players in the England league and about 50 german players, but Italy got 0 players in this league.
